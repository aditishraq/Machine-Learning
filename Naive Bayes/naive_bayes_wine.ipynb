```python
import pandas as pd
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
dir(wine)
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
df= pd.DataFrame(wine.data, columns=wine.feature_names)
df.head(20)
```

```python
df['target'] = wine.target
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis='columns'), wine.target, test_size=0.3)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
